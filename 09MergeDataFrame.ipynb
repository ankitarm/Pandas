```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df3 = pd.merge(df1,df2,on='city')
df3
```

```python
#database joins
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
#here it performs inner join
df3 = pd.merge(df1,df2,on='city')
df3
```

```python
df3 = pd.merge(df1,df2,on='city',how='inner')
df3
```

```python
df3 = pd.merge(df1,df2,on='city',how='outer')
df3
```

```python
df3 = pd.merge(df1,df2,on='city',how='left')
df3
```

```python
df3 = pd.merge(df1,df2,on='city',how='right')
df3
```

```python
#indicator for seeing from where the data came
df3 = pd.merge(df1,df2,on='city',how='outer',indicator=True)
df3
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3 = pd.merge(df1,df2,on='city')
df3
```

```python
df3 = pd.merge(df1,df2,on='city',suffixes=('_left','_right'))
df3
```

```python
df4 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df4.set_index('city',inplace=True)
df4
```

```python
df5 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df5.set_index('city',inplace=True)
df5
```
