```python
#changing string into date
df = pd.read_csv("weather_data2.csv", parse_dates=["day"])
df
```

```python
#type changed to date
type(df.day[0])
```

```python
#making date index
df.set_index('day',inplace=True)
```

```python
df
```

```python
#filling all NaN as 0
new_df = df.fillna(0)
new_df
```

```python
#setting different NaN values for different columns
new_df = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'No Event'
})
new_df
```

```python
#replacing 0 nan values with estimated values
#Carry forward Value
new_df = df.fillna(method='ffill')
new_df
```

```python
#Carry Backward Value
new_df = df.fillna(method='bfill')
new_df
```

```python
#backfilling column values/vertical coping
new_df = df.fillna(method='bfill',axis="columns")
new_df
```

```python
#carry forwarding only ones
new_df = df.fillna(method='ffill',limit=1)
new_df
```

```python
#carry forwarding two places
new_df = df.fillna(method='ffill',limit=1)
new_df
```

```python
#interpolate-for better estimation(comes up will middle value)
new_df = df.interpolate()
df,new_df
```

```python
#considering time also for estimation
new_df = df.interpolate(method='time')
df,new_df
```

```python
#dropping all rows with nan values
new_df = df.dropna()
new_df
```

```python
#drop  if all columns nan(dropped date 9 as all nan)
new_df = df.dropna(how='all')
new_df
```

```python
#dont drop if atleast 1 valid value in row
new_df = df.dropna(thresh=1)
df,new_df
```

```python
#dont drop if atleast 2 valid value(to keep row u need two valid value)
new_df = df.dropna(thresh=2)
df,new_df
```

```python
#reindexing based on dates
dt = pd.date_range("01-01-2017","01-11-2017")
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df
```
